### 矩形法
 
原理：使用近似的方法，求出若干小矩形的面积之和，用其逼近函数的面积。

近似方法：对于函数$f(x)$在区间$[a,b]$上的面积，对该区间N等分，每个子区间宽$\Delta x=\frac{b-a}{N}$。对每个子区间，取左端的端点$x_i$对应的函数值$f(x_i)$为高，画一个矩形，其中初始点$x_{0}=a$， 第i个子区间初始点$x_i=a+(i-1) \Delta x$。第i个子区间矩形的面积为$I_{i}=\Delta x f(x_i)$

所有矩形面积之和
  $I=\int _a^b\ f(x){dx} \approx \sum _{i=0}^{N-1}I_{i}=\sum _{i=0}^{N-1} \left(f (x_i)\right) \text{$\Delta $x}$

```python
# 使用python基本函数编程
import math

def rectangle(f,a,b,n):
    x0=a
    I=0
    deltax=float((b-a)/n)
    for i in range(n):

        I = I+f(x0)*deltax
        x0=x0+deltax

    return I


import matplotlib.pyplot as plt  
import numpy as np

a=0
b=np.pi/2
f=math.cos
print(rectangle(f,a,b,1),"误差为：",rectangle(f,a,b,1)-1)
print(rectangle(f,a,b,2),"误差为：",rectangle(f,a,b,2)-1)
print(rectangle(f,a,b,3),"误差为：",rectangle(f,a,b,3)-1)
print(rectangle(f,a,b,1000),"误差为：",rectangle(f,a,b,1000)-1)
# 绘制函数值如何随着分割的区间数N变化
x=np.arange(20)  
y=[rectangle(math.cos,a,b,i+1) for i in x ]
plt.plot(x,y,color="red",label="rect",linewidth=2) 
plt.ylim(0,2)
plt.legend(loc='lower right') 
plt.show()
```

```python
# 使用numpy计算

def np_rectangle(f,a,b,n):
    f = np.vectorize(f) # 使函数f可以对矢量进行运算， 即对输入的矢量X 中的每一个元素 x_i全部执行f(x_i)，并得到数组[f(x_1)...f(x_i)...]
    deltax=float((b-a)/n)
    x_list=np.linspace(a,b,n+1)
    I_list = deltax*f(x_list) # 这里由于linspace的分割方式，分N份，需要出现n+1个点
    return I_list.sum()

print(np_rectangle(f,a,b,1),"误差为：",np_rectangle(f,a,b,1)-1)
print(np_rectangle(f,a,b,2),"误差为：",np_rectangle(f,a,b,2)-1)
print(np_rectangle(f,a,b,3),"误差为：",np_rectangle(f,a,b,3)-1)
print(np_rectangle(f,a,b,1000),"误差为：",np_rectangle(f,a,b,1000)-1)
```

### 梯形法
每一个小区间梯形面积：$\text{I}_i=\frac{1}{2} \left(f (x_i)+f (x_{i+1})\right)\text{$\Delta $x}$

所有梯形面积之和：
\begin{eqnarray}
I&=&\sum_i I_i=\sum_i \frac{1}{2} \left(f (x_i)+f (x_{i+1})\right) \Delta x \\
 &=&\frac{1}{2}\left(f (x_0)+f (x_1)\right) \Delta x+ \frac{1}{2}\left(f (x_1)+f (x_2)\right) \Delta x\\
 &+&\text{...}+\frac{1}{2} \left(f (x_{N-2})+f (x_{N-1})\right)\Delta x \\
 &+&\frac{1}{2}\left(f (x_{N-1})+f (x_N)\right)\Delta x \\
 &=& \frac{1}{2} \left(f (x_0)\right) \Delta x+\left(f (x_1)+f (x_2)+\text{...}+f (x_{N-1})\right)\Delta x+\frac{1}{2}\left(f (x_N)\right) \Delta x
\end{eqnarray}
其中最后一步，由于相邻两个小梯形的公式中，均会出现$\frac{1}{2}f (x_i)$， 相加后变为$f (x_i)$。 但第一个和最后一个小梯形中的$\frac{1}{2} f (x_0)$和$\frac{1}{2} f (x_N)$无法与其他项合并。

因此最终的梯形面积公式为：
$$I=\frac{1}{2} \left(f (x_0)+f (x_N)\right) \Delta x+\left(f (x_1)+f (x_2)+\text{...}+f (x_{N-1})\right)\Delta x $$

```python
# 使用python基本函数编程
def trapezoid(f,a,b,n):
    x0=a
    deltax=(b-a)/n
    inte=(f(a)+f(b))/2*deltax
    for i in range(int(n)-1):
        x0=x0+deltax
        inte = inte+f(x0)*deltax
    return inte


import matplotlib.pyplot as plt  
import numpy as np

a=0
b=np.pi/2
f=np.cos
#输出不同n时的结果，以及误差
print(trapezoid(f,a,b,1),"误差为：",trapezoid(f,a,b,1)-1)
print(trapezoid(f,a,b,2),"误差为：",trapezoid(f,a,b,2)-1)
print(trapezoid(f,a,b,3),"误差为：",trapezoid(f,a,b,3)-1)
print(trapezoid(f,a,b,4),"误差为：",trapezoid(f,a,b,4)-1)
print(trapezoid(f,a,b,1000),"误差为：",trapezoid(f,a,b,1000)-1)

x=np.linspace(0,20,100) 
y=[trapezoid(f,a,b,i+1) for i in x ]
plt.plot(x,y,color="red",label="trapezoid",linewidth=2) 
plt.ylim(0,2)
plt.legend(loc='lower right') 
plt.show()
```

```python
# 使用numpy计算

def np_trapezoid(f,a,b,n):
    f = np.vectorize(f) # can apply on vector
    deltax = float(b - a)/n
    arr = f(np.linspace(a,b,n+1))
    return (deltax*(arr.sum() - arr[0]/2. - arr[-1]/2.)) # 这里减号的原因是 arr.sum()求和时，已经将第一项和最后一项求和了， 而本身公式中只要求对这两项求一半。

print(np_trapezoid(f,a,b,1000),"误差为：",np_trapezoid(f,a,b,1000)-1)
```

### 抛物线法（辛普森法）

课堂作业：自己推导迭代公式

f(x)是原函数，设$g(x)=Ax^2+Bx+C$是拟合后的函数。 将原函数的积分，换为抛物线函数的积分。

有
\begin{array}
f(a)=g(a)=Aa^2+Ba+C ,\\
f(b)=g(b)=Ab^2+Bb+C ,\\
f(\frac{(a+b)}{2}) = g(\frac{(a+b)}{2})=A\frac{(a+b)}{2}^2+B\frac{(a+b)}{2}+C$
\end{array}

积分时把$f(x)\sim g(x)$带入，可以得到：
\begin{array}{c}
\int_{a}^{b} f(x) d x \approx \int_{a}^{b} A x^{2}+B x+C d x \\
\int_{a}^{b} A x^{2}+B x+C d x=\frac{A}{3}\left(b^{3}-a^{3}\right)+\frac{B}{2}\left(b^{2}-a^{2}\right)+C(b-a)
\end{array}

其中：
\begin{array}{c}
b^{3}-a^{3}=(\mathrm{b}-\mathrm{a})\left(b^{2}+\mathrm{ab}+a^{2}\right) \\
b^{2}-a^{2}=(b-a)(b+a)
\end{array}
所以积分变为：
\begin{array}{c}
\int_{a}^{b} f(x) d x \sim  \frac{b-a}{6}\left( 2A\left(b^{2}+\mathrm{ab}+a^{2}\right)+3B\left(b+a\right)+6C)\right)
\end{array}
重新整理为：
\begin{eqnarray}{c}
\int_{a}^{b} f(x) d x &\sim&  \frac{b-a}{6}\left( (Aa^2+Ba+C)+(Ab^2+Bb+C)+4A(\frac{a+b}{2})^2+4B(\frac{a+b}{2})+4C\right)\\
&=& \frac{b-a}{6}\left( f(a) + f(b) + 4 f(\frac{a+b}{2}) + 4C\right)
\end{eqnarray}

以上针对区间[a,b], 故对每一个小区间[$x_i,x_{i+1}$]也都有：

$$\int_{x_i}^{x_{i+1}} f(x) d x=\frac{(x_{i+1}-x_i)}{6}[f(x_i)+f(x_{i+1})+4 f(\frac{(x_{i}+x_{i+1})}{2})]$$

将完整积分区间[a,b]分为N等分后， 求和结果为：
\begin{eqnarray}{c}
\int_{a}^{b} f(x) d x&=&\sum_{i=0}^{N}\int_{x_i}^{x_{i+1}}f(x) d x=\sum_{i=0}^{N-1}\frac{\Delta x}{6}[f(x_i)+f(x_{i+1})+4 f(\frac{(x_{i}+x_{i+1})}{2})]\\
\end{eqnarray}

```python
# 使用python基本函数编程
def simpson(f,a,b,n):
    deltax = float((b-a)/n)
    I= 0
    for i in range(n):
        xi=a+i*deltax
        xi1=a+(i+1)*deltax
        I+=1/6*deltax*(f(xi)+f(xi1)+4*f((xi+xi1)/2)) 
    return I
a=0
b=np.pi/2
f=np.cos
#输出不同n时的结果，以及误差
print(simpson(f,a,b,1),"误差为：",simpson(f,a,b,1)-1)
print(simpson(f,a,b,2),"误差为：",simpson(f,a,b,2)-1)
print(simpson(f,a,b,3),"误差为：",simpson(f,a,b,3)-1)
print(simpson(f,a,b,4),"误差为：",simpson(f,a,b,4)-1)
print(simpson(f,a,b,1000),"误差为：",simpson(f,a,b,1000)-1)
```

```python
### 使用numpy计算

def np_simpson(f,a,b,n):
    f = np.vectorize(f) # can apply on vector
    deltax = float(b - a)/n
    arr = f(np.linspace(a,b,n+1))
    arr_mid = f(np.linspace(a+deltax/2,b-deltax/2,n))
    return (deltax/6)*(2*arr.sum()- arr[0] - arr[-1] + arr_mid.sum()*4) # 这里减号的原因是 arr.sum()求和时，已经将第一项和最后一项求和了， 而本身公式中只要求对这两项求一半。

print(np_simpson(f,a,b,1000),"误差为：",np_simpson(f,a,b,1000)-1)
```

## 变步长积分
### 变步长梯形法
旧：  

$\text{T1}=\frac{1}{2} \Delta x \left(f\left(x_k\right)+f\left(x_{k+1}\right)\right)$

新：

取中点为 $\text x_{k+\frac{1}{2}}=\frac{1}{2} \left(x_k+x_{k+1}\right)$, 形成两个新的梯形为

$$\text{T2}=\frac{1}{4} \Delta x \left(f\left[x_k\right]+f\left[\frac{1}{2} \left(x_{k+1}+x_k\right)\right]+f\left[\frac{1}{2} \left(x_{k+1}+x_k\right)\right]+f\left[x_{k+1}\right]\right)$$

所以分为2n个梯形T2，但T2可以用T1的公式进行化简，用分为n个梯形的结果做化简，得到：

$\text{T2}=\frac{\text{T1}}{2}+\frac{1}{2} h f\left[\frac{1}{2} \left(x_k+x_{k+1}\right)\right]=\frac{\text{T1}}{2}+
\frac{1}{2} h f\left[x_{k+\frac{1}{2}}\right]$

把所有梯形累加

$T2_{n}=\frac{T1_{n}}{2}+\frac{1}{2} h \sum _{k=0}^{n-1} f\left[x_{k+\frac{1}{2}}\right]$

$ x_{k+\frac{1}{2}}=a+(k+\frac{1}{2}) \Delta x$

这个方法的好处在于，可以对格子不断细分，每次细分之后都可以利用之前的结果。

```python
# 用python自带函数编程

def trapezoid_vary(f,a,b,n):
    h=(b-a)
    int0=[(f(a)+f(b))*h/2]
    split=1/2
    for j in range(int(n)):
        split=split*2
        hh=h/split
        int1=0
        for i in range(int(split)):
            x0 = a+hh*i+hh/2
            int1=int1+f(x0)
        int2=int0[j]/2 + hh/2*int1
        int0.append(int2)
    return int0[-1]

n=2
a=0
b=np.pi/2
f=np.cos
print(trapezoid_vary(f,a,b,n),"误差为：",trapezoid_vary(f,a,b,n)-1)# 可以看到误差与直接梯形法中n=2时一致

import matplotlib.pyplot as plt  
import numpy as np

x=np.linspace(0,10,100) 
y=[trapezoid_vary(f,a,b,i+1) for i in x ]
plt.plot(x,y,color="red",label="trap1",linewidth=2) 
plt.ylim(0,2)
plt.legend(loc='lower right') 
plt.show()
```

```python
# 也可以直接调用之前写好的梯形法函数
def trapezoid_vary_iter(f,a,b,n):
    deltax = float(b - a)/n
    Tn= trapezoid(f,a,b,n)/2 
    for i in range(n):
        xi=a+(i+1/2)*deltax
        Tn+=deltax/2*f(xi)
    return(Tn)

a=0
b=np.pi/2
f=np.cos
n=2
print(trapezoid_vary_iter(f,a,b,n),"误差为：",trapezoid_vary_iter(f,a,b,n)-1)
```

### 变步长辛普森法
用 $ T_{n} $ 表示简单梯形法所求得的积分值。取$\left[x_{k}, x_{k+1}\right] $, 其中点坐标为  $x_{k+\frac{1}{2}}=\frac{1}{2}\left(x_{k}+x_{k+1}\right) $, 在该段上二分前和二分后的两 个积分值为  $T_{1}  和  T_{2} $, 它们分别可以表示为

\begin{array}{c}
T_{1}=\frac{h}{2}\left[f\left(x_{k}\right)+f\left(x_{k+1}\right)\right] \\
T_{2}=\frac{h}{4}\left[f\left(x_{k}\right)+2 f\left(x_{k+\frac{1}{2}}\right)+f\left(x_{k+1}\right)\right]
\end{array}

显然, 存在以下关系

$T_{2}=\frac{1}{2} T_{1}+\frac{h}{2} f\left(x_{k+\frac{1}{2}}\right)$

利用这个关系式, 当  k  从 0 到  n-1  累加时, 可导出下面的递推公式

$T_{2 n}=\frac{1}{2} T_{n}+\frac{h}{2} \sum_{k=0}^{n-1} f\left(x_{k+\frac{1}{2}}\right)$

因为$x_{k+\frac{1}{2}}=a+   \left(k+\frac{1}{2}\right) h  。$因此有

$T_{2 n}=\frac{1}{2} T_{n}+\frac{h}{2} \sum_{k=0}^{n-1} f\left[a+\left(k+\frac{1}{2}\right) h\right]$

与辛普森公式比较，即得到变步辛普森公式：
$$S_n=T_{2n}+\frac{1}{3}(T_{2n}-T_{n})$$

```python
def simpson_vary(f,a,b,n):
    Sn= trapezoid(f,a,b,2 *n) + (trapezoid(f,a,b,2*n)-trapezoid(f,a,b,n))/3
    return(Sn)

a=0
b=np.pi/2
f=np.cos
n=2
print(simpson_vary(f,a,b,n),"误差为：",simpson_vary(f,a,b,n)-1) #结果与直接辛普森法中n=2时结果一致，但运算比直接辛普森法更快速。
```

# 牛顿-科茨公式

如何理解以上的几个积分求解方法？

- 矩形法：$I=\int _a^b\text f(x){dx}\approx \sum _{i=0}^{N-1} \ f\left(x_i\right)\text{$\Delta $x} $

对于积分 $\int _a^b f(x)\text{dx} $ 使用直线$y=f\left(x_i\right)$来代替原始积分核$f(x)$

- 梯形法 使用斜线$a_0\text{+ }a_1x$ 代替$f[x]$, 得到梯形面积公式

回忆拉格朗日插值的过程， 这里就是使用插值拟合的公式$f(x)_{fit}$，来代替原有的f(x)。



例：对于两个点 $(x1,y1), (x2,y2)$

插值公式为：

$\text{l1=}\frac{x-\text{x2}}{\text{x1}-\text{x2}}$ 

$\text{l2=}\frac{x-\text{x1}}{\text{x2}-\text{x1}}$ 

$\phi (x)=\text{l1} \cdot \text{y1}+\text{l2}  \cdot \text{y2}$

整理为

$\text{$\phi $[x]=}\frac{1}{\text{x1}-\text{x2}}\text{y1} (x-\text{x2})-\text{y2} (x-\text{x1})= \frac{x (\text{y1}-\text{y2})}{\text{x1}-\text{x2}}+\frac{(\text{x1} \text{y2}-\text{x2} \text{y1})}{\text{x1}-\text{x2}}$

用$\phi$ 代替$f[x]$，积分变为

$\int _a^b  f(x)\text{dx}=\int _a^b \phi(x) \text{dx}$

对其积分，得到：

$\int _a^b \phi(x) \text{dx}=\frac{(\text{y1}-\text{y2})(\text{x1}-\text{x2})}{2}+(\text{x1} \text{y2}-\text{x2} \text{y1})$

整理即得到梯形面积公式：

$\text{dS}_i=\frac{1}{2} \left(f\left(x_{i+1}\right)+f\left(x_i\right)\right)\text{$\Delta $x} $


与梯形面积公式相同

### 一般方法 ：

使用拉格朗日多项式插值即可得到高次幂多项式$\phi(x)$代替$f(x)$. 



$ \phi (x)=\sum _k y_i l_i(x)$ , $l_i(x)=\prod _{j,j\neq i}\frac{x-x_j}{x_i-x_j}$

用 $\phi[x]$ 代替$f[x]$后，积分变为

$\int_a^b f[x] \, dx=
\int_a^b \phi [x] \, dx=\int _a^b\sum _k y_i l_i[x]dx =\sum _k y_i \int_a^b l_i[x] \, dx=\sum _k  y_iA[x]$

其中

$A(x)=\int_a^b l_i[x] \, dx=\int _a^b\prod _{j,j\neq i}\frac{\left(x-x_j\right)}{(x_i-x_j)} dx$

再将 $[a,b]$ 区间分为 n 份， $\text x_{k+1}=h l+x_k$对每一小份积分

$A(x)=\sum _{l=1}^n A_i $
