## sympy 执行常用数学计算

### 求解一元一次方程

$5*x+20=100$

要写成  ***=0 的形式，  用solve函数求解方程

```python
x = sy.symbols('x')#定义一个符号变量
answer=sy.solve([5*x+20-100],[x])#该函数的第一个参数是一个包含待解方程的列表，第二个参数是要解的变量列表
print (answer) # 不能省略乘号，solve() 函数会返回一个列表，其中包含解的值。
```

solve 是通用的解方程的函数，对于单变量方程，还可以使用solveset()

```python
sy.solveset(5*x+20-100,x)#solveset() 函数将返回一个集合，包含所有解。
```

```python
sy.solveset(x**2-1,x)#该函数的第一个参数是要解的方程，第二个参数是待解的变量。
```

### 求解二元一次方程组

$3*x+ 4*y=49 $

$8*x-y=14$

```python
x, y = sy.symbols('x y')
answer = sy.solve([3*x+ 4*y-49, 8*x-y-14],[x,y])#该函数的第一个参数是一个包含待解方程的列表，第二个参数是一个包含待解变量的列表。
print (answer) # 不能省略乘号
```

线性方程可以利用 linsolve() 求解.

```python
Eqns=[3*x+ 4*y-49, 8*x-y-14]#这行代码定义了一个包含两个方程的列表 Eqns。
sy.linsolve(Eqns,[x,y])#这行代码调用 sy.linsolve()函数来求解线性方程组。该函数的第一个参数是包含方程的列表，第二个参数是待解的变量列表。
```

利用方程组的解，可以进一步求其他函数

已知 $x + y =0.2, x + 3y = 1$  求 $x^2+4xy+4y^2$

```python
import sympy as sy
x, y = sy.symbols('x y')#定义符号变量
a = sy.solve([x + y -0.2, x + 3*y - 1],[x,y])#求解方程组
x= a[x]#提取方程组的解
y= a[y]#这两行代码从字典 a 中提取 x 和 y 的解，并重新赋值给 x 和 y。根据上一步的示例，x 和 y 将分别被赋值为 0.1。
formula= x**2 + 4*x*y + 4*y**2   # ** 代表乘方, 字符之间不能省略乘号
print(formula)
```

二重积分

$\int_{0}^{\pi} \int_0^x \frac{sin(t)}{\pi-t} dt dx$

```python
import sympy as sp

# 定义变量和函数
x, t = sp.symbols('x t')
expr = sp.sin(t)/ (sp.pi - t)

# 计算积分
result = sp.integrate(expr, (t, 0, x))  # 对 t 从 0 到 x 积分
result = sp.integrate(result, (x, 0, sp.pi))  # 对 x 从 0 到 pi 积分

# 打印结果
print("The value of the double integral is:", result)
```

$y'-3y=2$

其中我们知道 $y'$　一般是 $y'(t)=\frac{dy}{dt}$　的简写，　所以上式可以表达为

```python
#参考PPT
import sympy  as sy                                                                                        
t = sy.symbols('t')#定义符号变量
y = sy.Function('y')#定义函数:
#在 sy.dsolve 中，sy.Derivative 用来表示微分方程的导数项。
sy.dsolve(sy.Eq(sy.Derivative(y(t), t) -3*y(t), 2), y(t))
```

另一种写法
求初值问题

$f'(x)+f(x)+f^2(x)=0,~f(0)=1$

```python
import sympy  as sy  
f = sy.symbols('f', cls=sy.Function)#定义一个符号函数 f(x)，这里 cls=sy.Function 指定 f 是一个函数
x = sy.symbols('x')#定义了符号变量 x
eq = sy.Eq(f(x).diff(x,1)+f(x)+f(x)**2, 0)#定义了一个微分方程，使用 sy.Eq 来创建等式.f(x).diff(x, 1) 表示函数 f(x) 对 x 的一阶导数
con={f(0):1}#定义初始条件,这里定义了初始条件 con，表示在 x = 0 时，f(x) = 1。
exp=sy.dsolve(eq, f(x),ics=con)#sy.dsolve 函数来求解微分方程 eq，解的形式是关于 f(x) 的。使得解满足 f(0) = 1。
exp#输出结果
```

```python
求解微分，并对结果中初值导致的常数项进行代换，并画图：
```

```python
import sympy.plotting as syp #SymPy的绘图模块
import sympy as sy#导入SymPy库

t = sy.symbols('t')#定义一个符号变量t
y = sy.Function('y')#定义一个函数y

ex=sy.dsolve(sy.Eq(sy.Derivative(y(t), t)-y(t)+y(t)**2, 0), y(t))#求解微分方程，
print(ex)#这行代码打印微分方程的解。解的形式通常会包含一个积分常数 c1

ex1=ex.subs({sy.symbols("C1"):1}) #对常数C1进行代换,这行代码对解中的积分常数 C1 进行代换，将其设为1
print(ex1)#这行代码打印代换后的解。
syp.plot(ex1.rhs, (t,-4,4))#ex1.rhs 获取代换后解的右侧表达式，即解的具体形式,绘制解的图像，横轴范围为 -4到4
```

#### 约分 cancel()

```python
import sympy as sy

# 定义一个符号矩阵 M
M = sy.Matrix([[1, 2], [3, 4]])#定义矩阵M为一个2x2矩阵

# 计算 M 与 M 的逆矩阵的乘积，并打印其第一行第一列的元素
element = (M * M.inv())[0, 0]#矩阵M的逆矩阵M.inv()乘以M会得到单位矩阵，即[[1, 0], [0, 1]]
#[0, 0] 指的是取矩阵的第 0 行、第 0 列的元素。在单位矩阵中，这个位置的元素是 1。
print(element)  # 输出 1

# 简化第一行第一列的元素并打印
simplified_element = element.cancel() #虽然在这里cancel()方法没有实质性的效果，因为元素本身已经是最简形式的1。但如果元素是一个复杂的分数，cancel()方法会将其化简。
print(simplified_element)  # 输出 1
```

```python
import sympy as sy

# 定义符号
x = sy.symbols('x')

# 定义一个分数表达式
expr = (x**2 - 1) / (x**2 - x)

# 使用 .cancel() 方法简化
simplified_expr = expr.cancel()

# 打印结果
print(simplified_expr)
```

```python
import sympy as sy
import sympy.plotting as syp

# 定义符号变量
u = sy.symbols('u')

# 定义参数化曲线的 x 和 y 组件
xc = sy.cos(u) + sy.cos(7*u) / 2 + sy.cos(17*u) / 3
yc = sy.sin(u) + sy.sin(7*u) / 2 + sy.sin(17*u) / 3

# 绘制参数化曲线
syp.plot_parametric(xc, yc, (u, 0, 2 * sy.pi))#该函数接受三个参数：x 组件的表达式,y 组件的表达式,指定参数 u 的范围，从 0 到 2π。这将绘制一整圈
```

### 二维画图

```python
import sympy.plotting as syp#导入SymPy的绘图模块，并将其别名设置为syp。

x,y=sy.symbols("x y",real=True)#定义两个实数符号变量x和y

z=x+sy.I*y  #定义复数变量z，其中sy.I是SymPy表示虚数单位的符号。

w=sy.cos(z**2).expand(complex=True)#计算复数z的平方：z**2 = (x + iy)**2。然后计算cos(z**2)，这将返回一个复数表达式。expand(complex=True)：展开这个复数表达式
print(w)#输出计算得到的复数表达式w

wa=sy.Abs(w).expand(complex=True)#sy.Abs(w)：计算复数w的绝对值（模），展开绝对值表达式。

syp.plot_implicit(wa**2-1)#绘制|w|^2 = 1的图形，即模的平方等于1。plot_implicit()函数用于绘制隐式方程的图形
```

plot_implicit 是 SymPy 中用于绘制隐式方程图形的函数。隐式方程是指未显式地解出一个变量的方程，例如 F(x,y)=0
绘制单位圆的方程 x2+y2-1=0

#隐式方程是一种方程类型，其中变量没有明确地以一个变量的形式出现，而是与其他变量以某种方式隐含地相关

```python
import sympy as sy
import sympy.plotting as syp

# 定义符号变量
x, y = sy.symbols('x y')

# 定义隐式方程
equation = x**2 + y**2 - 1

# 使用 plot_implicit 绘制
syp.plot_implicit(equation, (x, -2, 2), (y, -2, 2), title='Unit Circle', xlabel='x', ylabel='y')
```
